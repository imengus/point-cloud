```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


mat = np.genfromtxt('ab.csv', delimiter=',')

%matplotlib inline

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(mat[:,0], mat[:,1], mat[:,2],  c=np.abs(mat[:,2])/np.max(mat), alpha=0.01, cmap='Greys')
ax.set_xlabel('x')
ax.view_init(elev=0, azim=180)
plt.show()
```
